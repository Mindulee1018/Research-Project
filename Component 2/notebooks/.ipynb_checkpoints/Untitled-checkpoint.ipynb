```python
import pandas as pd
import ast

# 1) Load (your file looks tab-separated)
df = pd.read_csv("SOLD_train.csv")

# 2) Parse rationales safely (they are stored like "[0,0,1]" or "[]")
def parse_list(x):
    if pd.isna(x):
        return []
    if isinstance(x, list):
        return x
    s = str(x).strip()
    try:
        return ast.literal_eval(s)
    except:
        return []

df["rationales_list"] = df["rationales"].apply(parse_list)

# 3) Extract hate words from tokens using rationale indices
def get_hate_words(tokens_str, r_list):
    toks = str(tokens_str).split()  # tokens are already space-separated
    n = min(len(toks), len(r_list))
    return " ".join([toks[i] for i in range(n) if int(r_list[i]) == 1])

df["hate_words"] = df.apply(lambda r: get_hate_words(r["tokens"], r["rationales_list"]), axis=1)

# 4) Create hate label from OFF/NOT
df["hate_label"] = df["label"].apply(lambda x: 1 if str(x).strip().upper() == "OFF" else 0)

# 5) (Optional) If you want hate_words only for OFF rows, clear it for NOT
df.loc[df["hate_label"] == 0, "hate_words"] = ""

# 6) Drop helper column and save
df.drop(columns=["rationales_list"], inplace=True)
df.to_csv("SOLD_train_with_hate.csv", sep="\t", index=False, encoding="utf-8")

df[["post_id", "tokens", "rationales", "label", "hate_words", "hate_label"]].head(10)

output_file = "SOLD_with_hate.tsv"
df.to_csv(output_file, sep="\t", index=False, encoding="utf-8")

print("Saved as:", output_file)
```

```python
import os
import math
import pandas as pd

df = pd.read_csv("SOLD_train_with_hate.csv", sep="\t", engine="python", encoding="utf-8")
df.head()

chunk_size = 200
out_dir = "SOLD_chunks_100"
os.makedirs(out_dir, exist_ok=True)

num_files = math.ceil(len(df) / chunk_size)

for i in range(num_files):
    start = i * chunk_size
    end = start + chunk_size
    chunk = df.iloc[start:end]
    
    file_path = os.path.join(out_dir, f"sold_part_{i+1:04d}.csv")
    chunk.to_csv(file_path, sep="\t", index=False, encoding="utf-8")

print(f"✅ Created {num_files} files in folder: {out_dir}")
```
