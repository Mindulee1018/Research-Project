```python
import pandas as pd
import ast

# Load dataset
df = pd.read_csv(r"C:\Users\fonse\SLIIT\projects\Research-Project\Component 2\data\SOLD_train.csv")

# Convert rationales from string to list
df["rationales"] = df["rationales"].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to extract hate words
def extract_hate_words(row):
    tokens = row["tokens"].split()          # split tokens
    rationales = row["rationales"]          # list of 0/1
    
    # Make sure lengths match
    if len(tokens) == len(rationales):
        hate_words = [token for token, r in zip(tokens, rationales) if r == 1]
        return hate_words
    else:
        return []

# Create new column with hate words
df["hate_words"] = df.apply(extract_hate_words, axis=1)

# Create binary hate column (1 if any hate word exists)
df["hate"] = df["hate_words"].apply(lambda x: 1 if len(x) > 0 else 0)

# View results
print(df[["tokens", "rationales", "hate_words", "hate"]].head())
```

```python
df.to_csv("SOLD_train_with_hate_column.csv", index=False)
```

```python
import os
import math
import pandas as pd

df = pd.read_csv(r"C:\Users\fonse\SLIIT\projects\Research-Project\Component 2\SOLD_train_with_hate_column.csv")
# Example: df.to_csv("SOLD_train_with_hate_column.csv", index=False)

batch_size = 200
out_dir = "batches_200"
os.makedirs(out_dir, exist_ok=True)

num_batches = math.ceil(len(df) / batch_size)

for i in range(num_batches):
    start = i * batch_size
    end = start + batch_size
    batch_df = df.iloc[start:end].copy()
    batch_df.to_csv(os.path.join(out_dir, f"batch_{i+1}.csv"), index=False)

print(f"✅ Done! Created {num_batches} files inside folder: {out_dir}")
```
